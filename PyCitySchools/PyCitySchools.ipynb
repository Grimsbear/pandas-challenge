```python
#Import dependencies
import pandas as pd
```

```python
#File paths to csv
schools = "Resources/schools_complete.csv"
students = "Resources/students_complete.csv"
```

```python
#Load csv files
schools_df = pd.read_csv(schools)
students_df = pd.read_csv(students)
#schools_df.head()
#students_df.head()
```

```python
#Merge csv files
merged_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
merged_df.head()
#merged_df.dtypes
```

```python
#Create District Summary
total_schools = len(pd.unique(merged_df["school_name"]))
#schools

total_students = len(merged_df["student_name"])
#student

total_budget = sum(pd.unique(merged_df["budget"]))
total_budget = "${:,.2f}".format(total_budget)
#total_budget

average_math = merged_df["math_score"].mean()
#average_math

average_read = merged_df["reading_score"].mean()
#average_read

passing_math_raw = merged_df["math_score"].gt(69)
#passing_math.sum()
passing_math = passing_math_raw.sum()/total_students
#passing_math

passing_read_raw = merged_df["reading_score"].gt(69)
#passing_read.sum()
passing_read = passing_read_raw.sum()/total_students
#passing_read

overall_pass_raw = merged_df["math_score"].gt(69) & merged_df["reading_score"].gt(69)
#overall_pass_raw.sum()
overall_pass = overall_pass_raw.sum()/total_students
#overall_pass
```

```python
#Create the actual DataFrame
district_sum = [{
    "Total Schools" : total_schools ,
    "Total Students" : total_students ,
    "Total Budget" : total_budget ,
    "Average Math Score" : average_math ,
    "Average Reading Score" : average_read ,
    "% Passing Math" : passing_math ,
    "% Passing Reading" : passing_read ,
    "% Overall Passing" : overall_pass
}]
district_df = pd.DataFrame(district_sum)
district_df.head()
```

```python
#Create School Summary
merged_df.rename(columns = {"school_name" : "School Name"},inplace = True)

school_raw = merged_df.groupby("School Name")
#school_raw.max()

school_type = school_raw["type"].max()
#school_type

school_students = school_raw["student_name"].count()
#school_students

school_budget = school_raw["budget"].max()
#school_budget

average_budget = school_budget/school_students
#average_budget

school_math_avg = school_raw["math_score"].mean()
#school_math_avg

school_read_avg = school_raw["reading_score"].mean()
#school_read_avg

school_passing_math_raw = merged_df[merged_df["math_score"] >= 70].groupby("School Name")["math_score"].count()
school_passing_math = (school_passing_math_raw/school_students) * 100
#school_passing_math_raw

school_passing_read_raw = merged_df[merged_df["reading_score"] >= 70].groupby("School Name")["reading_score"].count()
school_passing_read = (school_passing_read_raw/school_students) * 100
#school_passing_read_raw

overall_pass_school_raw = merged_df[merged_df["math_score"] >= 70].groupby("School Name")["math_score"].count() & merged_df[merged_df["reading_score"] >= 70].groupby("School Name")["reading_score"].count()
overall_pass_school = (overall_pass_school_raw/school_students) * 100
#overall_pass_school
```

```python
#Create School DataFrame
school_df = pd.DataFrame(school_type)
school_df["Total Students"] = school_students
school_df["Total Budget"] = school_budget
school_df["Average Budget per Kid"] = average_budget
school_df["Average Math Score"] = school_math_avg
school_df["Average Reading Score"] = school_read_avg
school_df["% Passing Math"] = school_passing_math
school_df["% Passing Reading"] = school_passing_read
school_df["% Overall Passing"] = overall_pass_school
school_df.rename(columns = {"type" : "School Type"}, inplace = True)
school_df.head(15)
```
